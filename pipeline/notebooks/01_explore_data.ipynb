```python
import sys
from pathlib import Path

# PROJECT ROOT = rag-pipeline/
PROJECT_ROOT = Path.cwd().parent            
SRC_DIR = PROJECT_ROOT / "src"

# add src to PYTHONPATH
sys.path.insert(0, str(SRC_DIR))

print("Project root:", PROJECT_ROOT)
print("Using src:", SRC_DIR)
import os
print(os.getcwd())
```

```python
from rag_pipeline.src.loaders import get_loader_for_path
```

```python



from cleaning import clean_text

# Notebooks live in rag-pipeline/notebooks, so go one level up
DATA_DIR = Path("../data/raw")  # adjust if your data lives elsewhere

if not DATA_DIR.exists():
    raise FileNotFoundError(f"Data dir not found: {DATA_DIR.resolve()}")

files = sorted(DATA_DIR.rglob("*"))
print(f"Found {len(files)} files under {DATA_DIR} (including dirs):\n")
for f in files:
    if f.is_file():
        print(" -", f.relative_to(DATA_DIR))

```

```python
# Pick a first supported file
supported_suffixes = {".txt", ".md", ".pdf", ".json", ".csv"}

candidates = [
    f for f in files
    if f.is_file() and f.suffix.lower() in supported_suffixes
]

if not candidates:
    raise RuntimeError("No supported files found (txt, md, pdf, json, csv).")

sample_path = candidates[0]
print(f"Using sample file: {sample_path}")

```

```python
# Load via loader and preview raw text
loader = get_loader_for_path(sample_path)
docs = loader.load(sample_path)

print(f"\nLoaded {len(docs)} Document(s)")

doc = docs[0]
print(f"Document ID: {doc.id}")
print(f"Metadata: {doc.metadata}")

print("\n--- RAW TEXT PREVIEW ---")
print(doc.text[:1000])

```

```python
# Cleaned preview
cleaned = clean_text(doc.text)

print("\n--- CLEANED TEXT PREVIEW ---")
print(cleaned[:1000])

```
